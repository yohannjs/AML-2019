    <h2>Dataset Description</h2>

* MSSubClass  
  * The building class
  * **Categorical data, 16 categories, categories should be converted to individual columns**
  * MUST BE CONVERTED
* MSZoning 
  * The general zoning classification
  * **Categorical data, 8 categories, categories should be converted to individual columns**
* LotFrontage 
  * Linear feet of street connected to property
  * Numerical data
* LotArea 
  * Lot size in square feet
  * Numerical data
* Street 
  * Type of road access
  * **Categorical data, 2 categories, categories should be converted to individual columns**
* Alley 
  * Type of alley access
  * **Categorical data, 3 categories, categories should be converted to individual columns**
* LotShape 
  * General shape of property
  * **Categorical data, 4 categories, categories can be converted, or be made numerical**
  * Reason to make numeric is because it is measured in 'degree of irregularity'
* LandContour 
  * Flatness of the property
  * **Categorical data, 4 categories, categories can be converted, or be made numerical**
  * Reason to make numeric is because it is measured in 'degree of flatness'  
* Utilities 
  * Type of utilities available
  * **Categorical data, 4 categories, categories can be converted, or be made numerical**
  * Reason to make numeric is because it is measured in 'number of utilities available'
* LotConfig 
  * Lot configuration
  * **Categorical data, 5 categories, categories should be converted to individual columns**
* LandSlope 
  * Slope of property
  * **Categorical data, 3 categories, categories can be converted, or be made numerical**
  * Reason to make numeric is because it is measured in 'degree of slope of property'
  * __probably correlated with LandContour__
* Neighborhood 
  * Physical locations within Ames city limits
  * **Categorical data, 25 categories, categories should be converted to individual columns**
* Condition1 
  * Proximity to main road or railroad
  * **Categorical data, 9 categories, categories should be converted to individual columns**
* Condition2 
  * Proximity to main road or railroad (if a second is present)
  * **Categorical data, 9/10 categories, categories should be converted to individual columns**
  * Not sure if 9 or 10 categories here
* BldgType 
  * Type of dwelling
  * **Categorical data, 5 categories, categories should be converted to individual columns**
* HouseStyle 
  * Style of dwelling
  * **Categorical data, 8 categories, categories should be converted to individual columns**
* OverallQual 
  * Overall material and finish quality
  * Numerical data
* OverallCond 
  * Overall condition rating
  * Numerical data
  * Possibly correlated with OverallQual
* YearBuilt 
  * Original construction date
  * Numerical data
* YearRemodAdd 
  * Remodel date
  * Numerical data
* RoofStyle 
  * Type of roof
  * **Categorical data, 6 categories, categories should be converted to individual columns**
* RoofMatl 
  * Roof material
  * **Categorical data, 8 categories, categories should be converted to individual columns**
* Exterior1st 
  * Exterior covering on house
  * **Categorical data, 17 categories, categories should be converted to individual columns**
* Exterior2nd 
  * Exterior covering on house (if more than one material)
  * **Categorical data, 17/18 categories, categories should be converted to individual columns**
  * Reason for uncertainty is because it is dependent on Exterior1st
* MasVnrType 
  * Masonry veneer type
  * **Categorical data, 5 categories, categories should be converted to individual columns**
* MasVnrArea 
  * Masonry veneer area in square feet
  * Numerical data
* ExterQualv 
  * Exterior material quality
  * **Categorical data, 5 categories, categories numerical**
* ExterCond 
  * Present condition of the material on the exterior
  * **Categorical data, 5 categories, categories numerical**
* Foundation 
  * Type of foundation
  * **Categorical data, 6 categories, categories should be converted to individual columns**
* BsmtQual 
  * Height of the basement
  * **Categorical data, 6 categories, categories numerical**
* BsmtCond 
  * General condition of the basement
  * **Categorical data, 6 categories, categories numerical**
* BsmtExposure 
  * Walkout or garden level basement walls
  * **Categorical data, 5 categories, categories can be converted, or be made numerical**
  * Could be made numerical because it measures 'degree of exposure of basement'
* BsmtFinType1 
  * Quality of basement finished area
  * **Categorical data, 7 categories, categories can be converted, or be made numerical**
  * Could be made numerical because it is a quantitive measure of quality of basement finish area 
* BsmtFinSF1 
  * Type 1 finished square feet
  * Numerical data
* BsmtFinType2 
  * Quality of second finished area (if present)
  * **Categorical data, 7/8 categories, categories can be converted, or be made numerical**
  * Could be made numerical because it is a quantitive measure of quality of basement finish area
  * Unsure on nr. of categories because it depends on BsmtFinType1
* BsmtFinSF2 
  * Type 2 finished square feet
  * Numerical data
* BsmtUnfSF 
  * Unfinished square feet of basement area
  * Numerical data
* TotalBsmtSF 
  * Total square feet of basement area
  * Numerical data 
  * Probably highly correlated with previous two measurements of basement square-feet area
* Heating

* HeatingQC

* CentralAir
